```python
import numpy as np
import xarray as xr
import uxarray as ux
import yaml
import cartopy
import holoviews as hv
import hvplot.pandas
from matplotlib import pyplot as plt, colors

# Set plot font sizes
plt.rcParams.update({'font.size': 16, 'axes.titlesize': 'medium'})

%matplotlib inline
```

```python
# Get paths
with open('../yaml/paths_LR.yaml', 'r') as f:
    paths = yaml.safe_load(f)
```

```python
bbox = [-100, 50, 39, 85]

uxda = ux.open_dataset(paths['meshfile'], paths['meshfile'])['bottomDepth'].subset.bounding_box(bbox[:2], bbox[2:])
gdf_data = uxda.to_geodataframe()
hv.extension('matplotlib')
projection = cartopy.crs.LambertConformal(-20, 70)

plot = gdf_data.hvplot.polygons(
    c='bottomDepth', cmap='Blues',
    colorbar=False, #clabel='Bottom depth [m]',
    width=6000, height=6000,
    projection=projection,
    rasterize=True,
)
# cbar_extend='neither', cbar_width=0.02
plot.opts(bgcolor='lightgray', norm='log', clim=(1, 5000), xlim=(-60, 10), ylim=(45, 80), title='Low Resolution (LR)')

fig = hv.render(plot)
fig.set_size_inches((8, 8))
```
